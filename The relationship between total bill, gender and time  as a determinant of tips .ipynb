```python
 # Continuous variable plot : Join plot, Disribution plot , Pair plot and Rug plot 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#(Univariate analysis) distribution plot is implimented over a continuous variable
```

```python
print(sns.distplot(tips['total_bill']))
```

```python
print(sns.distplot(tips['total_bill'],kde=False))# Setting the kde as False removes the kde line
```

```python
#print(sns.kdeplotlib(tips['total_bill']))# kdeplotlib  plots the kde
```

```python
print(sns.distplot(tips['total_bill'],kde=False, bins=30)) # using the bins to determine the numbers of bar plotted
```

```python

#  (Bivariate comparism) two continuous variable
```

```python
print(sns.jointplot(x='total_bill',y='tip',data=tips)) # comparing two continuous data
```

```python
print(sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')) 
```

```python
print(sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg'))
```

```python
print(sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde'))
```

```python
# Visualizing one variable against multiple variables
print(sns.pairplot(tips,hue='sex',palette='coolwarm'))
```

```python
print(sns.rugplot(tips['total_bill']))
```

```python
# Categorical variable plot: Bar plot, count plot, violin plot, stip plot and factor plot
```

```python
import numpy as np 
```

```python
print(sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std))
```

```python
print(sns.countplot(x='sex',data=tips))
```

```python
print(sns.boxplot(x='day',y='total_bill', data=tips))
```

```python
#'hue' are basically used to sepetrate a column into categories.
print(sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True))
```

```python
print(sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex',split=True))
```

```python
# The plot style combination 
print(sns.violinplot(x='day',y='total_bill',data=tips))
print(sns.swarmplot(x='day',y='total_bill',data=tips, color='black'))
```

```python
print (sns.factorplot(x='day',y='total_bill',data=tips,kind='bar'))
```

```python
#Heat Maps and cluster maps(using the flight manifest)
```

```python
# download the flight dataset( flight manifest)
flights = sns.load_dataset('flights')
print(flights.head(4))
```

```python
# Heat Maps and Styles
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')# .pivot_table() is to restructure the data
print(fp)
```

```python
sns.heatmap(fp,cmap='magma') 
```

```python
#linecolour and linewidths help to seperate the heatmap into visible readable grids
sns.heatmap(fp,cmap='magma',linecolor='white', linewidths=1)
```

```python
# Cluster mapping
```

```python
sns.clustermap(fp, cmap='coolwarm')#cmap is used to idetify the color clusters
```

```python
sns.clustermap(fp, cmap='coolwarm',standard_scale=1)#standard scale shows the intensity of the scale
```
